```python
from requests import get
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import requests

import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
url = 'https://www.airbnb.com.br/s/sao-paulo/homes'
response=requests.get(url,headers=headers)
soup=BeautifulSoup(response.content,'lxml')
```

```python
place_name = []
place_location = []
place_price = []
place_rating = []
place_comments = []
place_hosttype = []
place_rooms = []
place_facilities = []

listings_per_page = 20
pages_per_location = 15

for i in range(pages_per_location):
    offset = listings_per_page * i
    url = 'https://www.airbnb.com.br/s/sao-paulo/homes' + f'?items_offset={offset}'
    response=requests.get(url,headers=headers)
    soup=BeautifulSoup(response.content,'lxml')
    list_containers= soup.findAll(class_= '_8ssblpx')
    
    for item in range(len(list_containers)):
    
        place_name.append(list_containers[item].find(class_="_1whrsux9").text)
        place_location.append(list_containers[item].find(class_="_b14dlit").text)
        place_price.append(list_containers[item].find(class_="_krjbj").text[1:5])
        place_rooms.append(list_containers[item].find(class_="_kqh46o").text)
    
        try:
            place_facilities.append(list_containers[item].findAll(class_="_kqh46o")[1].text)
        except:
            place_facilities.append("Not Available")
    
        try:
            place_rating.append(list_containers[item].find(class_="_10fy1f8").text)
        except:
            place_rating.append("Not Available")
    
        try: 
            place_comments.append(list_containers[item].find(class_="_a7a5sx").text[2:5])
        except:
            place_comments.append("Not Available")
        
        try:
            place_hosttype.append(list_containers[item].find(class_="_ufoy4t").text)
        except:
            place_hosttype.append("REGULAR")
        

places_test = pd.DataFrame({
    'place_name': place_name,
    'place_location': place_location,
    'place_price': place_price,
    'place_rooms': place_rooms,
    'place_facilities': place_facilities,
    'place_rating': place_rating,
    'place_comments': place_comments,
    'place_hosttype': place_hosttype,
})

places_test.head(20)
```
